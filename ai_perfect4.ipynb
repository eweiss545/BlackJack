The confidence interval found will contain the true value of EG 95% of the time, and we can be even more precise if we drastically increase number of rounds played. And so in order to find the "theoretical" EG, we increased the number of rounds to 1 million (!) and found an EGgivenALL of -0.017739, within the confidence interval [-0.019739,-0.015739). This number is basically as close to the True EG as we can reasonably get and will be referred to as the True EG. The confidence interval is also extremely tiny, indicating that our True EG is very precise.

We also played a game (g1) of only 1,000 rounds. For this game, we get an EGgivenALL of 0.002800, within the confidence interval (-0.060570,0.066170). We will call this the measured EG, as 1,000 rounds is low enough to be considered experimental. Comparing the two EGs demonstrates that the measured EG can sometimes be considerably different from the True EG. The measured EG indicates a slightly positive gain, while the True EG indicates a slightly negative gain. While they only differ by around 0.01, the measured EG could be misleading because it indicates that PerfectNoMemory is a solid strategy (at least for breaking even). Looking at the confidence intervals, we can see that the True EG of -.017 does fall within the confidence interval of the measured EG, which makes sense, as the measured EG's confidence intervals are wider and in general the measured EG is far less precice due to lower sample size. The measured EG does not fall within the True EG's confidence intervals, which also makes sense, as the True EG's confidence intervals are much narrower and the True EG is much more precise. 

In terms of the strategy itself, AIPlayerPerfectNoMemory is not great because it results in a tiny loss, on average, per round played. Looking at equity curves, you can see that this player experiences a gradual decrease in bankroll after 1 million rounds played. The equity curve for the game with 1,000 rounds shows much more volatility, but on average, you can see a tendency towards a decrease in bankroll.
